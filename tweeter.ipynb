```python
from Scweet.scweet import scrape

elems = scrape(words=['책 추천'], since='2022-05-01', until='2022-05-09', \
    from_account = None, interval=1, headless=False, display_type='Top', \
    save_images=False, filter_replies=True)
```

```python
import pandas as pd

df = pd.read_csv("C:\jupiter_workspace\zam_project\outputs/책 추천_2022-05-01_2022-05-09.csv", encoding='utf-8')
df.head()
df.info()
```

```python
#df = df.drop(['UserScreenName', 'UserName', 'Text', 'Image link'], axis=1)
df['Comments'] = df['Comments'].fillna(0)
df['Likes'] = df['Likes'].fillna(0)
df['Retweets'] = df['Retweets'].fillna(0)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Timestamp'] = df['Timestamp'].dt.tz_localize(None)
df['Likes'] = df['Likes'].replace('천', '000')
df.insert(1, 'Weekday', df['Timestamp'].dt.day_name(), True)
df = df.astype({'Timestamp' : str, 'Weekday' : str, 'Comments' : int, 'Likes' : int, 'Retweets' : int})
df.head()
```

```python
df.to_csv('C:\jupiter_workspace\zam_project\outputs/책 추천_2022-05-01_2022-05-09.csv', encoding='utf-8')
```
